Iniciamo importando los módulos necesarios para trabajar la grafiación de datos.

https://matplotlib.org/tutorials/introductory/pyplot.html

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.units as munits
```

Para graficar, simplemente pasamos los datos a la función plot, y listo:

```python
plt.plot([1,2,3,4,5]) #Datos a graficar. Los valores en este caso serán para ambos ejes.
plt.ylabel('Eje Y') # Leyenda del eje Y
plt.xlabel('Eje X') # Leyenda del eje X
plt.show() # Da la orden para mostrar el gráfico diseñado previamente
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro') # Datos a graficar.
# Los valores en este caso son [X],[Y],'Estilo de formato para línea'
plt.ylabel('Eje Y') # Leyenda del eje Y
plt.xlabel('Eje X') # Leyenda del eje X
plt.axis([0,6,0,20]) # Define los límite para los ejes X y Y
plt.show() # Da la orden para mostrar el gráfico diseñado previamente
```

```python
# Intervalos de graficación cada 0.2 desde 0 hasta 5
t = np.arange(0,5,0.2)

# Graficación en masa dentro de un mismo gráfico
plt.plot(t, 2*t, 'r--', t, t**2, 'bs', t, (t/2)**3, 'g^')
plt.xlabel('eje X')
#plt.axis([-1,6, -1,11])
plt.show()
```
